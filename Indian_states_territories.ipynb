```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup as bs
```

```python
url = "https://en.wikipedia.org/wiki/States_and_union_territories_of_India"
```

```python
results = requests.get(url)
soup = bs(results.content, 'html.parser')
```

```python
states = soup.find_all("table") #wikitable sortable plainrowheaders jquery-tablesorter
indian_states = states[3]
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_html(str(indian_states))
df = pd.DataFrame(df[0])
df.to_csv('Indian_states.csv')

#df2 = pd.read_html(url)
#df2[3]
```

```python
df = pd.read_csv('Indian_states.csv')
```

```python
df.tail()
```

```python
df.rename(columns = {'Population(census 2011)[citation needed]':'Population(census 2011)', 'Officiallanguages[39]':'Officiallanguages', 'Additional officiallanguages[39]':'Additional officiallanguages'}, inplace = True)

df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace= True)
```

```python
df['Officiallanguages'] = df['Officiallanguages'].replace({r'[^\w\s]+':'',r'[\d$]':''}, regex=True)
df['Additional officiallanguages'] = df['Additional officiallanguages'].replace({r'[^\w\s]+':'',r'[\d$]':''}, regex=True)
df['Population(census 2011)'] = df['Population(census 2011)'].str.replace(r'[^\w\s]+', '')
df['Area(km2)'] = df['Area(km2)'].str.replace(r'[^\w\s]+', '')
```

```python
#error in data in row no. 23 in column Population(census 2011)[citation needed]	Area(km2)
population_23 = df['Population(census 2011)'][23] 
Area_23 = df['Area(km2)'][23]

df['Population(census 2011)'][23] = population_23[0:8]
df['Area(km2)'][23] = Area_23[:6]
# print(df['Population(census 2011)[citation needed]'][23])
# print(df['Area(km2)'][23])

df.tail()
```

```python
index = df[df['Additional officiallanguages'] == ''].index
df['Additional officiallanguages'][index] = df['Officiallanguages'][index]
df.drop
df.tail()
```
